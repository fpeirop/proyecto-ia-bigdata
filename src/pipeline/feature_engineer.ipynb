```python
import numpy as np
import pandas as pd
from reutilizable_functions import extract_lat_lon
from pathlib import Path
BASE_DIR = Path().resolve().parent.parent
```

```python
# Vamos a aplicar técnicas para de esta forma dejar el fichero procesado para que el modelo de ml lo pueda entender
traffic_df = pd.read_csv(f'{BASE_DIR}/data/raw/MTD_complete_data.csv')
```

```python
# Aplicamos one-hot-encoding para tener las variables categóricas en formato numérico
df_encoded = pd.get_dummies(traffic_df, columns=['day_type'], dtype=int)
df_encoded['date'] = pd.to_datetime(df_encoded['date'])
df_encoded['oneway'] = df_encoded['oneway'].astype(int)
```

```python
# Aplicamos one-hot-encoding para tener las variables categóricas en formato numérico
df_encoded = pd.get_dummies(df_encoded, columns=['highway'], dtype=int)
```

```python
# Con esta forma vamos a conseguir la latitud y longitud en columnas separadas, todo enfocado al modelo de ml
original_coords = extract_lat_lon(df_encoded, "original_point")
closest_coords = extract_lat_lon(df_encoded, "closest_point")
df_encoded.drop(columns=['original_point', 'closest_point'], inplace=True)
```

```python
# Aquí concatenamos el df anterior con las coordenadas
df = pd.concat([df_encoded, original_coords, closest_coords], axis=1)
file_path = Path(f'{BASE_DIR}/data/processed/MTD_complete_data.csv')

if not file_path.exists():
    df.to_csv(file_path, index=False)
else:
    print("El archivo ya existe, no se sobrescribe")
```
